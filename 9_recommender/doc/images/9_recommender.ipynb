```python
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

figsize = (5, 3)

warnings.filterwarnings('ignore')

IMAGES_PATH = '.'
os.makedirs(IMAGES_PATH, exist_ok=True)
def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=500):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
# Generating a power-law distribution
x = np.linspace(1, 100, 100)
y = x**(-.8) # exponent is chosen arbitrarily for illustration

# Plotting the figure with more visible area under the curve
plt.figure(figsize=figsize)
# Plotting the power-law distribution with thicker lines
plt.plot(x, y, linewidth=2, color='green')
# Filling the area under the curve with brighter colors and no transparency
plt.fill_between(x, y, where=(x<=10), color='green', label='Népszerű termékek')
plt.fill_between(x, y, where=(x>10) & (x<=60), color='orange', label='Némileg népszerű termékek')
plt.fill_between(x, y, where=(x>60), color='yellow', label='Niche termékek')
# Adding title and labels
plt.title('Hosszú farok eloszlás')
plt.xlabel('Egyedi termékek száma')
plt.ylabel('Népszerűség')
# Adding the legend with a larger font
plt.legend()
# Show the plot
save_fig('recommender_1')
plt.show()
```
