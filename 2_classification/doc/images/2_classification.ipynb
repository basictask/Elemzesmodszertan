```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "fundamentals"
IMAGES_PATH = "."
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
marker_size = 100

def plot_binary_classification(ax):    
    ax.scatter([1, 2], [1, 2], color='blue', label='1. Osztály', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='2. Osztály', s=marker_size)
    ax.set_title('Bináris osztályozás')
    ax.legend()

def plot_multiclass_classification(ax):
    ax.scatter([1, 2, 3], [1, 2, 3], color='blue', label='1. Osztály', s=marker_size)
    ax.scatter([1.5, 2.5, 3.5], [1, 2, 3], color='red', label='2. Osztály', s=marker_size)
    ax.scatter([2, 3, 4], [1.5, 2.5, 3.5], color='green', label='3. Osztály', s=marker_size)
    ax.set_title('Multiosztályos osztályozás')
    ax.legend()

def plot_multilabel_classification(ax):
    ax.scatter([1, 2], [1, 2], color='blue', label='1. Osztály', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='2. Osztály', marker='x', s=marker_size)
    ax.set_title('Multicímkés osztályozás')
    ax.legend()

def plot_multioutput_classification(ax):
    ax.scatter([1, 2], [1, 2], color='blue', label='Output 1 (1. Osztály)', s=marker_size)
    ax.scatter([1, 2], [1.5, 2.5], color='green', label='Output 1 (2. Osztály)', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='Output 2 (1. Osztály)', marker='x', s=marker_size)
    ax.scatter([2, 3], [2.5, 3.5], color='orange', label='Output 2 (2. Osztály)', marker='x', s=marker_size)
    ax.set_title('Multioutput osztályozás')
    ax.legend()



fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_binary_classification(ax)
save_fig('osztalyozas_7')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multiclass_classification(ax)
save_fig('osztalyozas_8')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multilabel_classification(ax)
save_fig('osztalyozas_9')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multioutput_classification(ax)
save_fig('osztalyozas_10')
plt.show()
```

```python
x = [200, -200, 300]
y = [0, 1, 0]

plt.figure(figsize=(5,3))
plt.scatter(x, y)
plt.title('Nyereségek sakkmeccsen')
plt.ylabel('Fehér nyert?')
plt.xlabel('Játékosok rangjának különbsége')
plt.xlim([min(x) - 100, max(x) + 100])
plt.ylim([-.25, 1.25])
save_fig('osztalyozas_12')
plt.show()
```

```python
x = np.array(x).reshape(-1, 1)  # Features
y = np.array(y)                # Target variable

# Fit the linear regression model
model = LinearRegression()
model.fit(x, y)

# Making predictions for plotting
x_fit = np.linspace(x.min() - 100, x.max() + 100, 1000).reshape(-1, 1)
y_fit = model.predict(x_fit)

plt.figure(figsize=(5,3))
plt.scatter(x, y)
plt.plot(x_fit, y_fit, color='red', label='Modell')
plt.title('Nyereségek sakkmeccsen')
plt.ylabel('Fehér nyert?')
plt.xlabel('Játékosok rangjának különbsége')
plt.ylim([-.25, 1.25])
plt.xlim([x.min() - 100, x.max() + 100])
save_fig('osztalyozas_13')
plt.show()
```
