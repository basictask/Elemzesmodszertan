```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "fundamentals"
IMAGES_PATH = "."
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
marker_size = 100

def plot_binary_classification(ax):    
    ax.scatter([1, 2], [1, 2], color='blue', label='1. Osztly', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='2. Osztly', s=marker_size)
    ax.set_title('Binris osztlyozs')
    ax.legend()

def plot_multiclass_classification(ax):
    ax.scatter([1, 2, 3], [1, 2, 3], color='blue', label='1. Osztly', s=marker_size)
    ax.scatter([1.5, 2.5, 3.5], [1, 2, 3], color='red', label='2. Osztly', s=marker_size)
    ax.scatter([2, 3, 4], [1.5, 2.5, 3.5], color='green', label='3. Osztly', s=marker_size)
    ax.set_title('Multiosztlyos osztlyozs')
    ax.legend()

def plot_multilabel_classification(ax):
    ax.scatter([1, 2], [1, 2], color='blue', label='1. Osztly', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='2. Osztly', marker='x', s=marker_size)
    ax.set_title('Multicmks osztlyozs')
    ax.legend()

def plot_multioutput_classification(ax):
    ax.scatter([1, 2], [1, 2], color='blue', label='Output 1 (1. Osztly)', s=marker_size)
    ax.scatter([1, 2], [1.5, 2.5], color='green', label='Output 1 (2. Osztly)', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='Output 2 (1. Osztly)', marker='x', s=marker_size)
    ax.scatter([2, 3], [2.5, 3.5], color='orange', label='Output 2 (2. Osztly)', marker='x', s=marker_size)
    ax.set_title('Multioutput osztlyozs')
    ax.legend()



fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_binary_classification(ax)
save_fig('osztalyozas_7')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multiclass_classification(ax)
save_fig('osztalyozas_8')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multilabel_classification(ax)
save_fig('osztalyozas_9')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multioutput_classification(ax)
save_fig('osztalyozas_10')
plt.show()
```
