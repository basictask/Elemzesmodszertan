```python
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "fundamentals"
IMAGES_PATH = "."
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
marker_size = 100

def plot_binary_classification(ax):    
    ax.scatter([1, 2], [1, 2], color='blue', label='1. Osztly', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='2. Osztly', s=marker_size)
    ax.set_title('Binris osztlyozs')
    ax.legend()

def plot_multiclass_classification(ax):
    ax.scatter([1, 2, 3], [1, 2, 3], color='blue', label='1. Osztly', s=marker_size)
    ax.scatter([1.5, 2.5, 3.5], [1, 2, 3], color='red', label='2. Osztly', s=marker_size)
    ax.scatter([2, 3, 4], [1.5, 2.5, 3.5], color='green', label='3. Osztly', s=marker_size)
    ax.set_title('Multiosztlyos osztlyozs')
    ax.legend()

def plot_multilabel_classification(ax):
    ax.scatter([1, 2], [1, 2], color='blue', label='1. Osztly', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='2. Osztly', marker='x', s=marker_size)
    ax.set_title('Multicmks osztlyozs')
    ax.legend()

def plot_multioutput_classification(ax):
    ax.scatter([1, 2], [1, 2], color='blue', label='Output 1 (1. Osztly)', s=marker_size)
    ax.scatter([1, 2], [1.5, 2.5], color='green', label='Output 1 (2. Osztly)', s=marker_size)
    ax.scatter([2, 3], [2, 3], color='red', label='Output 2 (1. Osztly)', marker='x', s=marker_size)
    ax.scatter([2, 3], [2.5, 3.5], color='orange', label='Output 2 (2. Osztly)', marker='x', s=marker_size)
    ax.set_title('Multioutput osztlyozs')
    ax.legend()



fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_binary_classification(ax)
save_fig('osztalyozas_7')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multiclass_classification(ax)
save_fig('osztalyozas_8')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multilabel_classification(ax)
save_fig('osztalyozas_9')
plt.show()

fig = plt.figure(figsize=(5,3))
ax = fig.add_subplot(1,1,1)
plot_multioutput_classification(ax)
save_fig('osztalyozas_10')
plt.show()
```

```python
x = [200, -200, 300]
y = [0, 1, 0]

plt.figure(figsize=(5,3))
plt.scatter(x, y)
plt.title('Nyeresgek sakkmeccsen')
plt.ylabel('Fehr nyert?')
plt.xlabel('Jtkosok rangjnak klnbsge')
plt.xlim([min(x) - 100, max(x) + 100])
plt.ylim([-.25, 1.25])
save_fig('osztalyozas_12')
plt.show()
```

```python
x = np.array(x).reshape(-1, 1)  # Features
y = np.array(y)                # Target variable

# Fit the linear regression model
model = LinearRegression()
model.fit(x, y)

# Making predictions for plotting
x_fit = np.linspace(x.min() - 100, x.max() + 100, 1000).reshape(-1, 1)
y_fit = model.predict(x_fit)

plt.figure(figsize=(5,3))
plt.scatter(x, y)
plt.plot(x_fit, y_fit, color='red', label='Modell')
plt.title('Nyeresgek sakkmeccsen')
plt.ylabel('Fehr nyert?')
plt.xlabel('Jtkosok rangjnak klnbsge')
plt.ylim([-.25, 1.25])
plt.xlim([x.min() - 100, x.max() + 100])
save_fig('osztalyozas_13')
plt.show()
```

## Logisztikus regresszi

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())

X = iris["data"][:, 3:]  # petal width
y = (iris["target"] == 2).astype(np.int)  # 1 if Iris virginica, else 0
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver="lbfgs", random_state=42)
log_reg.fit(X, y)
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize=(8, 3))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Nem Iris virginica")
plt.text(decision_boundary+0.02, 0.15, "Dntsi hatr", fontsize=14, color="k", ha="center")
plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.xlabel("Szirom hossz (cm)", fontsize=14)
plt.ylabel("Becslt valsznsg", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])
save_fig("osztalyozas_20")
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.int)

log_reg = LogisticRegression(solver="lbfgs", C=10**10, random_state=42)
log_reg.fit(X, y)

x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]

y_proba = log_reg.predict_proba(X_new)

plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)


left_right = np.array([2.9, 7])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]

plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.text(3.5, 1.5, "Nem Iris virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Szirom hossz", fontsize=14)
plt.ylabel("Szirom szlessg", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])
save_fig("osztalyozas_21")
plt.show()
```

```python
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs", C=10, random_state=42)
softmax_reg.fit(X, y)

x0, x1 = np.meshgrid(
        np.linspace(0, 8, 500).reshape(-1, 1),
        np.linspace(0, 3.5, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]


y_proba = softmax_reg.predict_proba(X_new)
y_predict = softmax_reg.predict(X_new)

zz1 = y_proba[:, 1].reshape(x0.shape)
zz = y_predict.reshape(x0.shape)

plt.figure(figsize=(10, 4))
plt.plot(X[y==2, 0], X[y==2, 1], "g^", label="Iris virginica")
plt.plot(X[y==1, 0], X[y==1, 1], "bs", label="Iris versicolor")
plt.plot(X[y==0, 0], X[y==0, 1], "yo", label="Iris setosa")

from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])

plt.contourf(x0, x1, zz, cmap=custom_cmap)
contour = plt.contour(x0, x1, zz1, cmap=plt.cm.brg)
plt.clabel(contour, inline=1, fontsize=12)
plt.xlabel("Szirom hossz", fontsize=14)
plt.ylabel("Szirom szlessg", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 7, 0, 3.5])
save_fig("osztalyozas_22")
plt.show()
```
