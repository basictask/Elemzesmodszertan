```python
import os
import numpy as np
import matplotlib.pyplot as plt

IMAGES_PATH = '.'
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=500):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
# Generate more realistic sample data for episodic rewards
episodes = np.arange(1, 101)  # 100 episodes
np.random.seed(0)
rewards = np.log(episodes) + np.random.normal(scale=0.5, size=episodes.shape)

# Create the plot with the updated data
plt.figure(figsize=(5, 3))
plt.plot(episodes, rewards, label='Epizodikus jutalom')
plt.xlabel('Epiz√≥d')
plt.ylabel('Jutalom')
plt.legend()
save_fig('reinforcement_10')
plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_eps_decay(eps_start, eps_end, eps_decay, n_iter):
    eps = eps_start
    track = [eps]
    for _ in range(n_iter):
        eps = eps * eps_decay if eps > eps_end else eps_end
        track.append(eps)
    plt.figure(figsize=(5, 5))
    plt.plot(track)
    plt.xlabel('Epiz√≥d')
    plt.ylabel(r'Felfedez√©si val√≥sz√≠n≈±s√©g ($\varepsilon$)')
    save_fig('reinforcement_15')
    plt.show()


max_i = 50000
eps_end = 0  # Felfedez√©si r√°ta v√©gs≈ë √©rt√©ke
eps_decay = 0.9999  # Felfedez√©si r√°ta p√°rologtat√°si egy√ºtthat√≥ja
eps_start = 1.0  # Felfedez√©si r√°ta kezd≈ë√©rt√©ke
eps = eps_start

plot_eps_decay(eps_start, eps_end, eps_decay, max_i)
```
